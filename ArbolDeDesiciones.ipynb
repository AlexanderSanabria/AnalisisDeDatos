## 4. Dividir el dataset en Train y Test para evaluar correctamente el modelo

```python
# Porcentaje de datos para entrenamiento (train) = 75% , y para prueba (test) = 25%


#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size=0.25, random_state= 13)
```

```python
print (X_train)
```

```python
print (Y_train)
```

## 5. Entrenar el modelo configurando los diferentes hiperparámetros

```python

tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, Y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
pd.crosstab(np.array([y[0] for y in Y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(Y_test,predicciones)
accuracy
```

## 6. Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc

```python
print(classification_report(Y_test,predicciones))
```

```python
accuracy = accuracy_score(Y_test,predicciones)
accuracy
```

## 7. Realizar las diferentes gráficas que permitan visualizar los resultados del modelo

```python
#Matrix de Confusión del modelo

cm=confusion_matrix(Y_test,predicciones)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

## 8. Interpretar, analizar y documentar los resultados obtenidos

    La precisión indica la proporción de observaciones clasificadas como positivas que fueron clasificadas correctamente. En este caso, las clases 4, 5, 6 y 7 tienen precisiones que van del 0.50 al 0.70, lo que sugiere que el modelo es relativamente bueno para predecir estas clases.
    Recall,  indica la proporción de observaciones positivas que fueron correctamente identificadas por el modelo. Nuevamente, las clases 4, 5, 6 y 7 tienen valores razonables de recall.
    F1-score es la media armónica de precisión y recall, y proporciona una medida balanceada entre ambas métricas. En general, las clases 4, 5, 6 y 7 tienen F1-scores moderados.

    Este comportamiento lo podemos atribuir a que para las diferentes calidades de vino la mayoria de los datos entan centrados en calidades 5 y 6
